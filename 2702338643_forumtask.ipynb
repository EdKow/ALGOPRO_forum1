```python
lloyd = {
    "name": "Lloyd",
    "homework": [90.0,97.0,75.0,92.0],
    "quizzes": [88.0,40.0,94.0],
    "tests": [75.0,90.0]
}

alice = {
    "name": "Alice",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0]
}

tyler = {
    "name": "Tyler",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0]
}

students = [lloyd, alice, tyler]

for student in students:
    print(f"Student's name         : {student['name']}")
    print(f"Student's homework     : {student['homework']}")
    print(f"Student's quizzes      : {student['quizzes']}")
    print(f"Student's tests        : {student['tests']}")
    print(" ")
    
def average(numbers):
    total = float(sum(numbers))
    avg = total / len(numbers)
    return avg

def get_average(student):
    homework = average(student['homework'])
    quizzes = average(student['quizzes'])
    tests = average(student['tests'])
    totalAvg = (homework*0.1) + (quizzes*0.3) + (tests*0.6)
    return totalAvg

def lettergrade(score):
    if score >= 0 and score <= 100:
        if score >= 90:
            return "A"
        elif score >= 80:
            return "B"
        elif score >= 70:
            return "C"
        elif score >= 60:
            return "D"
        else:
            return "F"
    else:
        return "Invalid Score"

    
def classaverage(students):
    results = []
    for student in students:
        results.append(get_average(student))
    avg = average(results)
    return avg
    
print(f"Class Average                : {classaverage(students)}")
print(f"Class's Average (Rounded)    : {round(classaverage(students), 2)}")
print(f"Class's Average letter grade : {lettergrade(classaverage(students))}")
```

```python
week = ["sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

money = []

for days in week:
    sales = eval(input(f'Enter the sales for {days}: '))

    while sales < 0:
        sales = eval(input(f'Input was invald. Re-Enter the sales for {days}: '))
    else:
        money.append(sales)

print(f'Total sales is             \t: ${sum(money):,.2f}')
print(f'The minimum sale amount is \t: ${min(money):,.2f}')
print(f'The maximum sale amount is \t: ${max(money):,.2f}')


```
